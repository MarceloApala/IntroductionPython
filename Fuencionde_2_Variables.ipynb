```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x1=np.arange(0.0, 2.0, 0.01)
y1 = np.sin(2 * np.pi * x1)     

fig, ax = plt.subplots()
ax.plot(x1, y1)

ax.grid(True, linestyle='-.')
ax.tick_params(labelcolor='r', labelsize='medium', width=3)

plt.show()                #Grafica Curvas
```

```python
x1=np.arange(-4,4.01,0.01)#DOminio funciones ,extremo final abierto
x2=np.arange(-4,4.01,0.01)
y1=x1**4-4*x1**2+2*x1+3       #Funciones 
y2=-x2**4+4*x2**2-2*x2-3

plt.figure(figsize=(6,6))  #Tamano de la grafica
plt.xlim(-3,3)            #Limites de los ejes
plt.ylim(-10,10)
plt.xticks(np.arange(-3,4,1))#Configurar ejes ,extremo final abierto
plt.yticks(np.arange(-10,11,1))

plt.plot(x1,y1,'ko-')         #Grafica funciones y estilo como color grosor
plt.plot(x2,y2,'ko-')

idx=np.argwhere(np.diff(np.sign(y1-y2))).flatten() #Calc Intersecciones
plt.plot(x1[idx],y1[idx],'bo') #Grafica de intersecciones 

plt.title('Funciones Polinomicas de 4to grado')
plt.grid()                #Grafica reticulado
plt.show()                #Grafica Curvas

print('Coords intersecciones x : ',x1[idx])
print('Coords intersecciones x : ',y1[idx]) #Imprime intersecciones
```

```python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)


fig, (ax1, ax2) = plt.subplots(2, sharey=True)

ax1.plot(x1, y1, 'ko-')
ax1.set(title='A tale of 2 subplots', ylabel='Damped oscillation')

ax2.plot(x2, y2, 'r.-')
ax2.set(xlabel='time (s)', ylabel='Undamped')

plt.show()
```
