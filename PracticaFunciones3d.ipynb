```python
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import cm
from matplotlib.ticker import LinearLocator
import numpy as np
```

```python
from mpl_toolkits.mplo3d import Axes3D
mpl.rcParams["figure.figsize"]=[6.4,4.8]

ax=plt.axes(projection='3d')

X = np.arange(-10,10.25,0.25)
Y = np.arange(-8,8.25,0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)



surf = ax.plot_surface(X, Y, Z,rstride=2,cstride=2, cmap=plt.cm.get_cmap('Blues', 6),
                       linewidth=0.2, edgecolor='b')

ax.set_zlim(-10,10)
ax.zaxis.set_major_locator(LinearLocator(11))

ax.set_zlim(-6,6)
ax.zaxis.set_major_locator(LinearLocator(7))

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig=plt.figure()
ax=Axes3D(fig)

X = np.arange(-4,4.2,0.2)
Y = np.arange(-4,4.2,0.2)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = 2+np.sin(R)



surf = ax.plot_surface(X, Y, Z,rstride=1,cstride=1, cmap=plt.cm.get_cmap('cubehelix', 6),
                       linewidth=0.2, edgecolor='w')

ax.contour(X,Y,Z,zdir='z',offset=-3,cmap=plt.cm.get_cmap('cubehelix',6))

ax.set_zlim(-3,3)

plt.show()
```

```python
fig=plt.figure()
ax=plt.axes(projection='3d')

x,y=np.meshgrid(np.linspace(-1,1,30),np.linspace(-1,1,30))
z=np.cos(x*np.pi)*np.sin(y*np.pi)

ax.plot_wireframe(x,y,z,rstride=1,cstride=1,linewidth=0.5,color='black')

plt.show()
```

```python
ax=plt.axes(projection='3d')

X = np.arange(-10,10.5,0.5)
Y = np.arange(-8,8.5,0.5)
X, Y = np.meshgrid(X, Y)
Z = 6-2*X-3*Y

ax.plot_wireframe(X,Y,Z,cmap='jet',linewidth=0.2,edgecolor='black')

plt.show()
```

```python
mpl.rcParams['legend.fontsize']=10

fig=plt.figure()
ax=fig.gca(projection='3d')

theta=np.linspace(-4*np.pi,4*np.pi,100)

z=np.linspace(-2,2,100)
r=z**2+1
x=r*np.sin(theta)
y=r*np.cos(theta)

ax.plot(x,y,z ,label='parametric curve')
ax.legend()

ax.set_xlabel('$X$',fontsize=10,rotation=150)
ax.set_ylabel('$Y$')
ax.set_zlabel(r'$\gamma$',fontsize=10,rotation=60)
ax.yaxis._axinfo['label']['space_factor']=3.0

plt.show()
```
