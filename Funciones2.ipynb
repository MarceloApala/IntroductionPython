```python
import numpy as np
import matplotlib.pyplot as plt 
```

**FUNCIONES**

**a) Rectas,parabolas,intersecciones**

```python
x1=np.arange(-4,12.01,0.01)#DOminio funciones ,extremo final abierto
x2=np.arange(-4,12.01,0.01)
y1=0.25*x1**2-2*x1+2       #Funciones 
y2=-0.5*x2+6

plt.figure(figsize=(6,7))  #Tamano de la grafica
plt.xlim(-6,14)            #Limites de los ejes
plt.ylim(-6,16)
plt.xticks(np.arange(-6,15,1))#Configurar ejes ,extremo final abierto
plt.yticks(np.arange(-6,17,1))

plt.plot(x1,y1,'y')         #Grafica funciones y estilo como color grosor
plt.plot(x2,y2,'r--')

idx=np.argwhere(np.diff(np.sign(y1-y2))).flatten() #Calc Intersecciones
plt.plot(x1[idx],y1[idx],'bo') #Grafica de intersecciones 

plt.title('Rectas,parabolas,intersecciones')
plt.grid()                #Grafica reticulado
plt.show()                #Grafica Curvas

print('Coords intersecciones x : ',x1[idx])
print('Coords intersecciones x : ',y1[idx]) #Imprime intersecciones
```

**Funciones Polinomicas**

```python
x1=np.arange(-4,4.01,0.01)#DOminio funciones ,extremo final abierto
x2=np.arange(-4,4.01,0.01)
y1=x1**4-4*x1**2+2*x1+3       #Funciones 
y2=-x2**4+4*x2**2-2*x2-3

plt.figure(figsize=(6,6))  #Tamano de la grafica
plt.xlim(-3,3)            #Limites de los ejes
plt.ylim(-10,10)
plt.xticks(np.arange(-3,4,1))#Configurar ejes ,extremo final abierto
plt.yticks(np.arange(-10,11,1))

plt.plot(x1,y1,'y')         #Grafica funciones y estilo como color grosor
plt.plot(x2,y2,'r--')

idx=np.argwhere(np.diff(np.sign(y1-y2))).flatten() #Calc Intersecciones
plt.plot(x1[idx],y1[idx],'bo') #Grafica de intersecciones 

plt.title('Funciones Polinomicas de 4to grado')
plt.grid()                #Grafica reticulado
plt.show()                #Grafica Curvas

print('Coords intersecciones x : ',x1[idx])
print('Coords intersecciones x : ',y1[idx]) #Imprime intersecciones
```

```python
from numpy import sin,pi
x1=np.arange(-14,14.01,0.01)#DOminio funciones ,extremo final abierto
x2=np.arange(-14,14.01,0.01)
y1=sin(x1)      #Funciones 
y2=3*sin(1.0*(x2-2))+0

plt.figure(figsize=(8,4))  #Tamano de la grafica
plt.xlim(-14,14)            #Limites de los ejes
plt.ylim(-5,5)
plt.xticks(np.arange(-14,16,2))#Configurar ejes ,extremo final abierto
plt.yticks(np.arange(-5,6,1))

plt.plot(x1,y1,'y')         #Grafica funciones y estilo como color grosor
plt.plot(x2,y2,'r--')

plt.title('Funciones seno')
plt.grid()                #Grafica reticulado
plt.show()                #Grafica Curvas

```
