```python
from sympy import Symbol,factor,expand,solve
x=Symbol('x')
y=Symbol('y')
z=Symbol('z')
```

**Sistema de Ecuaciones**

**Ejercicio 1**

1.jpg

```python
expr=2*x-3*y-9
expr1=4*x+3*y-9
res=solve((expr,expr1))
print('Resultado =',res )
```

**Ejercicio 2**

**Ejercicio 3**

```python
expr=x+2*y-7
expr1=5*x-y-2
res=solve((expr,expr1))
print('Resultado = ',res)
```

**Ejercicio 4**

4.jpg

```python
expr=x-2*y+3*z-1
expr1=x+2*y-z-13
expr2=3*x+2*y-5*z-3
res=solve((expr,expr1,expr2))
print('Resultado = ',res)
```

*Importar libreria para operar desigualdades*
