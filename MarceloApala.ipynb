```python
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib import cm
from matplotlib.ticker import LinearLocator
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

**Ejercicio 1**

```python
x1=np.arange(-6,6.01,0.01)
x2=np.arange(6,12.01,0.01)
x3=np.arange(0,12.01,0.01)
x4=np.arange(-6,0.01,0.01)
y1=(1/8)*x1**2-(1/2)*x1+(5/2)
y2=-(1/4)*x2**2+4*x2-11
y3=-(1/8)*x3**2+2*x3-5

plt.figure(figsize=(6,7)) 
plt.xlim(-6,14)            
plt.ylim(-6,16)
plt.xticks(np.arange(-7,14,1))
plt.yticks(np.arange(-6,14,1))

plt.plot(x1,y1,'r')
plt.plot(x2,y2,'r')
plt.plot(x3,y3,'b')

plt.grid()
plt.show()
```

**Ejercicio 2**

```python

mpl.rcParams["figure.figsize"]=[6.4,4.8]

ax=plt.axes(projection='3d')

X = np.arange(-2,2.05,0.1)
Y = np.arange(-2,2.05,0.1)
X, Y = np.meshgrid(X, Y)
R = 0.1+X*np.exp(-X**2-Y**2)#np.sqrt(X**2 + Y**2)
Z = np.sin(R)



ax.plot_surface(X, Y, Z,rstride=1,cstride=1, cmap=plt.cm.get_cmap('cubehelix', 6),
                       linewidth=0.2, edgecolor='w')

ax.contour(X,Y,Z,zdir='z',offset=-3,cmap=plt.cm.get_cmap('cubehelix',6))

ax.set_zlim(-0.5,0.5)
ax.zaxis.set_major_locator(LinearLocator(11))

plt.grid()
plt.show()
```
